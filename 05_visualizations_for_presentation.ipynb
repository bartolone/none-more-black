```python
# These arrays are taken from the confusion matrices in my modeling notebook.

genres = ['rock', 'experimental', 'electronic', 'hip-hop', 'folk', 'pop', 'instrumental', 'international', 'classical']

totals = np.array([1464, 1085, 839, 323, 299, 204, 309, 128, 87])
base_correct = np.array([1143, 770, 567, 117, 29, 9, 75, 21, 18])
oversamp_correct = np.array([1039, 738, 524, 147, 68, 35, 102, 44, 29])
top3_correct = [1184, 780, 603]

base_pct = base_correct / totals
oversamp_pct = oversamp_correct / totals
top3_pct = top3_correct / totals[0:3]
print(base_pct)
print(oversamp_pct)
print(top3_pct)
top3_pctn = np.array([0.809, 0.719, 0.719, 0, 0, 0, 0, 0, 0])
print(top3_pctn)
```

```python
import matplotlib.pyplot as plt

SMALL_SIZE = 10
MEDIUM_SIZE = 16
BIGGER_SIZE = 20

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=BIGGER_SIZE)    # legend fontsize
plt.rc('axes', titlesize=BIGGER_SIZE)

f, ax = plt.subplots(figsize=(12,8))
plt.bar(genres, base_pct)
plt.ylabel('Accuracy')
plt.xlabel('Genre')
plt.title('Per-Genre Accuracy for Base Model\n(58% Overall Accuracy)')
plt.ylim([0,1])

#plt.show()

plt.savefig('base.png', dpi=800)
```
